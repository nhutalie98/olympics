```python
medals = pd.read_csv('olympic_medals.csv', index_col= None)
```

```python
medals['Gold'] = medals.Medal == 'Gold'
medals['Silver'] = medals.Medal == 'Silver'
medals['Bronze'] = medals.Medal == 'Bronze'
```

```python
# There are so many countries in teh dataset that it becomes messy to plot visulizations comparing the countries
# To make this more neat, we will create subsets of the data for North American, European, and Asian countries
north_america = medals[medals["Country"].isin(["United States", "Mexico", "Canada", "Guatemala", "Cuba", "Haiti",
                                              "Dominican Republic", "Honduras", "El Salvador", "Nicaragua", "Costa Rica",
                                              "Panama", "Jamaica", "Trinidad and Tobago", "Bahamas", "Belize", "Barbados",
                                              "Saint Lucia", "Saint Vincent and the Grenadines", "Grenada", "Antigua and Barbuda",
                                              "Dominica", "Saint Kitts and Nevis"])]

europe = medals[medals["Country"].isin(["Russia", "Germany", "United Kingdom", "France", "Italy", "Spain", "Ukraine", "Poland",
                                       "Romania", "Netherlands", "Belgium", "Czech Republic", "Greece", "Portugal", "Sweden",
                                       "Hungary", "Belarus", "Austria", "Serbia", "Switzerland", "Bulgaria", "Denmark", "Finland",
                                       "Slovakia", "Norway", "Ireland", "Croatia", "Moldova", "Bosnia and Herzegovina", "Albania",
                                       "Lithuania", "North Macedonia", "Slovenia", "Latvia", "Estonia", "Montenegro", "Luxembourg",
                                       "Malta", "Iceland", "Andorra", "Monaco", "Liechtenstein", "San Morino", "Holy See"])]

asia = medals[medals['Country'].isin(['China', 'India', 'Indonesia', 'Pakistan', 'Bangladesh', 'Japan', 'Phillipines', 'Vietnam',
                                     'Turkey', 'Iran', 'Thailand', 'Myanmar', 'South Korea', 'Iraq', 'Afghanistan', 'Saudi Arabia',
                                     'Uzbekistan', 'Malaysia', 'Yemen', 'Nepal', 'North Korea', 'Sri Lanka', 'Kazakhstan', 'Syria',
                                     'Cambodia', 'Jordan', 'Azerbaijan', 'United Arab Emirates', 'Tajikistan', 'Israel', 'Laos',
                                     'Lebanon', 'Kyrgystan', 'Turkmenistan', 'Singapore', 'Oman', 'Kuwait', 'Georgia', 'Mongolia',
                                     'Armenia', 'Qatar', 'Bahrain', 'Timor-Leste', 'Cyprus', 'Bhutan', 'Maldives', 'Brunei'])]


north_america.head()
europe.head()
asia.head()
```

```python
# Population of each country

na_pop = sns.catplot(data=north_america, kind="bar", x="Country Code", y="Population")
plt.title('North American Country Populations')
na_pop.fig.set_figwidth(15)
na_pop.fig.set_figheight(5)

eur_pop = sns.catplot(data=europe, kind='bar', x='Country Code', y='Population')
plt.title('European Country Populations')
eur_pop.fig.set_figwidth(15)
eur_pop.fig.set_figheight(5)

asia_pop = sns.catplot(data=asia, kind='bar', x='Country Code', y='Population')
plt.title('Asian Country Populations')
asia_pop.fig.set_figwidth(15)
asia_pop.fig.set_figheight(5)
```

* We split up the data into three separate regions. These regions were North America, Europe, and Asia. The graphs above explore the population difference between countries in each of these reagions. As we can see The United States has the highest population in North America followed by Mexico. In Europe Russia has the highest population followed by Germany, France, Great Britain, and Italy.In Asia China has the highest population followed closely by Inida.



```python
# Medals won by each Country
# Split by type of medal (gold/silver/bronze)

na_medals = sns.catplot(data=north_america, kind="count", x="Country Code", hue='Medal', palette='pastel')
plt.title('North America Olympic Medals')
na_medals.fig.set_figwidth(15)
na_medals.fig.set_figheight(5)

eur_medals = sns.catplot(data=europe, kind="count", x="Country Code", hue='Medal', palette='pastel')
plt.title('Europe Olympic Medals')
eur_medals.fig.set_figwidth(15)
eur_medals.fig.set_figheight(5)

asia_medals = sns.catplot(data=asia, kind="count", x="Country Code", hue='Medal', palette='pastel')
plt.title('Asia Olympic Medals')
asia_medals.fig.set_figwidth(15)
asia_medals.fig.set_figheight(5)
```

* In the graphs above we can see the medals won by each country in the three different regions. In the North American region the United States has by far the most medals won. Canada is second in medals won even though they have a smaller population compared to Mexico. In Europe Great Britain has the most medals won. In Asia the biggest difference between population and medals won is India. India has the second highest population but the third highest number of medals won behind Japan and China. It is interesing to note that population does not exactly correlate to the number of medals a country will win.



```python
# Medals won by each country
# Split by games (summer/winter)

na_medals_games = sns.catplot(data=north_america, kind="count", x="Country Code", hue='Games', palette='pastel')
plt.title('North America Olympic Medals by Games')
na_medals_games.fig.set_figwidth(15)
na_medals_games.fig.set_figheight(5)

eur_medals_games = sns.catplot(data=europe, kind="count", x="Country Code", hue='Games', palette='pastel')
plt.title('Europe Olympic Medals by Games')
eur_medals_games.fig.set_figwidth(15)
eur_medals_games.fig.set_figheight(5)

asia_medals_games = sns.catplot(data=asia, kind="count", x="Country Code", hue='Games', palette='pastel')
plt.title('Asia Olympic Medals by Games')
asia_medals_games.fig.set_figwidth(15)
asia_medals_games.fig.set_figheight(5)
```

* In the graphs above the number of medals won in either the summer/winter olympics for each country is displayed. In North America the United States has considerably more medals won in the summer then the winter olympics.
```python
# Relationship between Number of Medals and Country's GDP per Capita
# First we create the table to count Medal, groupby Country
# Then we create another table to select unique value from Country and GDP per Capita column from the data frame
med = medals.groupby("Country")["Medal"].count()
med = pd.DataFrame(med)
gdp = medals[["Country","GDP per Capita"]].drop_duplicates()
med
gdp
```

```python
# We then merge the two tables created above by inner join
df_c = pd.merge(left = med, right = gdp, left_on='Country',right_on='Country',how='inner')
df_c
```

```python
# Finally, based on the merged table, we graph linear regression between Total Medals earned by Countries and GDP per Capita of each Country
corr = sns.lmplot(x="Medal", y="GDP per Capita", data=df_c)
corr.fig.set_figwidth(18)
plt.title("Linear regression between Total Medals earned by Countries and GDP per Capita")
```

```python
# Correlation statistic between Total Medals earned by Countries and GDP per Capita
pg.corr(x=df_c['Medal'], y=df_c['GDP per Capita'])
```

* Based on the graph and statistic table, there is not quite strong correlation (r = 0.44) between GDP per Capita and Total Medals earned

```python
# GDP per Capita of each continent
# First find unique value of each country per capita on each continent created by subsetting data frame
# Second we calculate sum, median, min, max GDP per Capita 
# Then we create data frame table for the statistics we have calculated above
am = north_america[["Country","GDP per Capita"]].drop_duplicates()
am_gdp = am['GDP per Capita'].sum()
am_median = am['GDP per Capita'].median()
am_min = am['GDP per Capita'].min()
am_max = am['GDP per Capita'].max()

aa = asia[["Country","GDP per Capita"]].drop_duplicates()
aa_gdp = aa['GDP per Capita'].sum()
aa_median = aa['GDP per Capita'].median()
aa_min = aa['GDP per Capita'].min()
aa_max = aa['GDP per Capita'].max()

eu = europe[["Country","GDP per Capita"]].drop_duplicates()
eu_gdp = eu['GDP per Capita'].sum()
eu_median = eu['GDP per Capita'].median()
eu_min = eu['GDP per Capita'].min()
eu_max = eu['GDP per Capita'].max()

cv = [('America',am_gdp, am_median, am_min, am_max),('Asia', aa_gdp, aa_median, aa_min, aa_max), ('Europe', eu_gdp, eu_median, eu_min, eu_max)]
continent = pd.DataFrame(cv, columns=['Continent', 'Sum GDP/Capita','Median GDP/Capita','Min','Max'])
#pd.options.display.float_format = "{:.2f}".format
#continent.describe()
continent
```
